<h1 style="text-align:center;color:blue;">Keyboard Shortcuts</h1>

```
# File
ctrl + s : Save
ctrl + Shift + S : Save As (PDF / Template)

ctrl + p : print
Alt + WI : Page Preview
Alt + WL : Normal Page Veiw
Alt + PI : Page Print Setting

# Navigation
Enter : Cell Down         ; Shift + Enter : Cell Up
Tab   : To the right side ; Shift + Tab   : Cell Left
Alt + Enter : Next row withing the cell

Arrows : Nevigate Arrows
ctrl + Arrows : Go to the last cell in the data
Ctrl + Shift + Arrow : Select upto last cell in the data
Ctrl + Shift + * : Select All data
ctrl + Home : Back to A1

Ctrl + A : Select all sheet
Ctrl + Space : Select Column  ; Ctrl + Shift + (+) / Alt + HIC --> Add Column     ; Ctrl - --> Delete Cell
Shift + Space : Select Row    ; Ctrl + Shift + (+) / Alt + HIR --> Add Row Above  ; Ctrl - --> Delete Cell

# Hide Rows & Column
Ctrl + 0  : Hide Column   |   Ctrl  + Space -> Alt + AG/AU : Group/Ungroup
Ctrl + 1  : Hide Rows     |   Shift + Space -> Alt + AG/AU : Group/Ungroup 

# Adding / Removing Cell
Ctrl + Shift + (+) : Add Cell   ; Ctrl + Shift + (+) : Insert cut column to left to selected column
Ctrl + (-)  : Delete the cell
del       : Remove content
Alt + HEF : Remove Format
Alt + HEA : Remove All


# Sheet
Alt + HOR : Rename the sheet
Shift + F11 : Add Sheet to right
Alt + HDS : Delete the sheet
Alt + EM : Move or Copy sheet
Ctrl + PgDown/PgUp : Navigating throught sheets
Ctrl + DrgagSheet : Copy sheet
Ctrl + Shift + F6 : Get a list of active sheet

# Important Functions
Alt + = + Enter : Auto Sum
```

<h1 style="text-align:center;color:blue;">Functions</h1>

```
There are 461 build in function in Excel
Syntax --> =FUNCTIONNAME(ARGUMENTS)
```

```
# Keyboard shortcut
shift + f3 : Opening Function Argument window
f2 : Formula bar activation

# Regular Function
SUM()
MIN()
MAX()
AVERAGE()
COUNT()    : It does not count Text and Empty Space
SUBTOTAL() : Does Change Its Value as Filter Applied

# DataBase Functions (Operation base on some criteria)
DSUM()
DAVERAGE()
DCOUNT()


# Conditional Formula
AND()
OR()

# if - else
IF()
COUNTIF()
SUMIF()
IFERROR()

# Excel Lookup Functions (selecting data base on index and columns names)
VLOOKUP()    : Give Index Vertically (Column Index in table range) : Lookup value should be in first column
HLOOKUP()    : Give Index Horizontally (Row Index in the table range) : Lookup Value should be in first row

XLOOKUP()

INDEX()      : Returns the value at specific position
MATCH()      : Return the position
INDEX( data(f4) ,
       MATCH( maching_index_cell(f43),   indexes(f4), 0 ) ,
       MATCH( matching_column_cell(f42), columns(f4), 0 )
     )


# Text base functions
LEN()
LEFT()
RIGHT()
MID()
SEARCH()  : Retn the position of letter in a text
CONCATENATE()
```
