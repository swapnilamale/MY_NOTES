###  File Permissions
```
1.chmod
2.chown
3.chgrp
4.getfacl / setfacl
```

```
- Linux is multi-user OS
- It works on premise that "Every user has responsibility for controlling access to their files"
- Every file directory created by you can be protected from , made accessible to other users 
  by changing the access permissions

# There are 3 types of permissions (rwx)
r - read
w - write (Read/Modify/delete)
x - execute (Running a program)
    (in case of directory x means, you use cd command for that directory)
    

# These Permissions actually give at 3 levels
u - User
g - Group (can be people in the same project)
o - Other (public : everyone on the system) 
a - all


# chmod (change mode)
chmod g-w <file_name>  : "remove","write" permission from "group" for "<file_name>"
chmod a-w <file_name>  : "remove","write" permission from "all"   for "<filename>"
chmod u-w <file_name>  : "remove","write" permission from "group" for "<file_name>"
                          In this case even thogh you want to delete the file, you cannont
                          only root account can delete this file now

chmod o+r  <file_name> : "Add", "read"    permission for "everyone" for "<file_name>"
chmod u+rw <file_name> : "Add","read and write" permission for "user" for "<file_name>"

| permission using numbers |
--- : 0
--x : 1
-w- : 2
-wx : 3
r-- : 4
r-x : 5
rw- : 6
rwx : 7

chmod 777 <file_name> : Give all the permission at all level



# Change the Ownership of the Permissions
- There are two owners of file & directory
     1. user : root useraccount is defualt
     2. group
  You can change both user & group (you cannont change root account (Administrator) as it owns everything)
  
  
chown  <user> <file_name> : change owner to <user> for <file_name>
chgrp -R <group> <file_name> : change group to <grp> for <file_name>




# ACL (Access Control List)
Lets have a scenario in which a particular user is not a member of group created by you but still want to give permission to him explicitly without adding him to any group!
Basically, ACCL are used to make flexible permission Mechanism in Linux


getfacl <file_name>    : Displaying Permission of the File

setfacl -m u:<user>:rwx <file_name>   : Add permission to user
setfacl -dm u:<user>:rwx <folder_name> : Add Permission to directory
setfacl -x u:<user> <file_name>      : Remove Permition for perticular user
setfacl -dx u:<user> <file_name>
setfacl -b  <file_name>   : Make permition to default / reset permition

m : modifiy ; x: remove 
u : user

Note: 
1. As you assign ACL permission to file/directory it adds + sign at the end of the permission (ls -l)
2. Setting w permission in ACL: does not allow to remove a file in case of ACL
```

##### Special Permissions 

```
There are 3 additional special permission in linux
setuid     :  This permition allows user to run the command where as it will be executed by root in background
setgid     :  This permition allows group to run the command where as it will be executed by root in the background
sticky bit :  Permiton that allow only user and root to delete thoes files (- --- --- --t)
              any user cannot delete that directory
              Ex : /temp


chmod u+s <file>   : assign special permission at user level                    (- --s --- --- )
chmod g+s <file>   : assign special permission at group level                   (- --- --s --- )
chmod +t <file>    : adding sticky bit to the file                              (- --- --- --t )
                     other than user and root , anyone could write the file     
                     will not not be permitted to delete the file/directories or subdeirectories

chmod u-s <file>   :  remove special permition at user level
chmod g-s <file>   :  remove special permition at group level


find all the commands with special permition (setuid / setgid)
find / -perm /6000 -type f
```

###  Linux Account Authentication

```
Lets consider an enterprise level scenario where there are 1000 of servers
Now when new users come, for him/her it is very overwhelming to create account in each server.
it become further more complex when you want to update user info or change the password for each server.
In Corperate,Things dont work like that

For that Every Operating sys has Account Authentication Programm (Directory Service)
Ex:
Windows has "Active Directory" Program
In this you create an account and your info will be stored in the user-directory-database as user
& whenever you want to login to linux server account authentication goes against the user-directory-database, while logging in server will communicate with "Active Directory" using communicator "WinBIND", gives authentication.
Fetch user information and user can log in, in that server

This process of communicating with "Account Authentication Programm" , "Active Directory" in our case
is executed using protocal called LDAP (Lightweight Directory Acess Protocal)


Note : if you want to login any server, you will need SSH Protocal
       if you want to login to any server using "Active Directory" , you will need LDAP protocal
       

Directory Service for different OS
Windows : Active Directory
Linux   : IDM (Identity Manager) , OpenLDAP (Note : LDAP is a protocol)
```

#####  Proccess Mangaement in a Nutshell

```
Running Proccess in background,foreground & after exit terminal
Killing the proccess
Proccess priority
proccess monitoring
list proccess

Note : Whenever you run the proccess by default
       1. You will not get promt back untill compleation of proccess
       2. proccess attached himself to the terminal session i.e you close the terminal
          proccess will be terminated
       3. whatever info,error or warnings messages are there associated with proccess will
          be sent to screen 
       4. linux by default assign proccess priority 0 to proccess
          
- Running proccess in background    : (Ctrl + z) --> jobs --> bg
- bring it proccess into foreground : fg
- Running procces after quiting terminal:
  to run the proccess irrespective of the terminal we use "nohup"
  
  nohup <proccess command> &                    : Now this proccess is NOT attached 
                                                  to the terminal
  nohup <proccess command> & > /dev/null 2>&1 & : whatever the info message, error or 
                                                  warnig message are there, send it to the
                                                  /dev/null dont send it on screen
                                                  
- Priotizing Proccess:
  proccess priority scale of linux vary between (-20 - 19)
  lower the proccess priority scale highest will be the priority of the proccess
  (-20 : Highest Priority, 19: Lowest Priority)
  linux by default assign proccess priority 0 to the process if not specified
  you can change proccess priority by command "nice"
  
  nice -n <process_priority_scale> <process command>
    
- List process 
  ps -ef 

- Process Monitoring
  top
  
- killing process
  kill / pkill
```

IP  :  Internet Protocol
       A unique string of numbers sepearated by periods that identifies each computer using the Internet Protocol to 
       communicate over a network
       
Subnet mask : Associate way to attach IP address       
              32 bit number that "masks" an IP address, and divides the IP address into network address & host address
              it is made by setting network bits to all "1"'s and setting host bits to all "0"'s
           
Static : IP Address not changed while rebooted
DHCP : IP Address does change while rebooted

```

```
Gateway  : Which route you have to pick for traffic input/output
           for Home Environment IP address of Modem will be gateway
```

```
Interface : Port on your device where you connect a cable
            Interface has "Interface MAC address" & this MAC Address Never changes
```


##### Network Files & Commands

```
