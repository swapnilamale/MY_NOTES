````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
df=pd.DataFrame({'x':np.linspace(0,5,11) , 'y':np.linspace(0,5,11)**2})
df.head()
````

<div style="text-align: center;">
    <h1> Simple Plot </u> </h1>
</div>

````python
plt.figure(figsize=(8,6),dpi=70)                    # Sizes in inches
plt.plot(df.x,df.y,'r-o',lw=2,alpha=0.5,label='y')
plt.legend(loc=0,frameon=False)                     # Select Best location

plt.suptitle('PicBoo',fontsize=25)
plt.title('X vs Y',fontsize=12)

plt.xlabel('X',fontsize=10)
plt.ylabel('Y',fontsize=10)

plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

plt.xlim([1,4])
plt.ylim([0,20])

plt.grid(color='b', alpha=0.5, linestyle='dashed', lw=0.9)

plt.text(3.5,12.5,"this is the lable i can add")
````

````python
df.plot.line(x='x',y='y',figsize=(5,3),lw=2,marker='o',ls='--',color='r')
plt.legend(loc=(-.15,0.9))
````

````python
df.plot.area(alpha=0.5)
plt.legend(ncol=2)
````

````python
df.plot.bar(stacked=False)
````

````python
plt.plot(df.x,'r-o',label='x')
plt.plot(df.y,'b-o',label='y')
plt.legend(loc=9)

plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
````

````python
fig=plt.figure( figsize=(5,3) , dpi=70 )

ax=fig.add_axes([0,0,1,1])
ax.plot(df.x,'r-o',label='X')
ax.plot(df.y,'b-o',label='Y')
ax.legend(loc=10)

ax.set_title('X AND Y')
ax.set_xlabel('NA')
ax.set_ylabel('X / Y')
````

    <h1> Multiple Axes Plot </u> </h1>

````python
fig=plt.figure(figsize=(7,5),dpi=70 )

ax1=fig.add_axes([0,0,1,1])
ax1.plot(df.x,df.y,'r-o',label='y')
ax1.legend(loc=4)
ax1.set_title('X vs Y')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2=fig.add_axes([0.08,0.65,0.4,0.3])
ax2.plot(df.y,df.x,'b-o',label='x')
ax2.legend(loc=0)
ax2.set_title('Y vs X')
ax2.set_xlabel('Y')
ax2.set_ylabel('X')
````

````python
fig = plt.figure(figsize=(5,3))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
````

````python
fig,ax= plt.subplots(1,2,figsize=(7,2.5),dpi=70)

ax[0].scatter(df.x,df.y,label='y')
ax[0].legend()
ax[0].set_title('X vs Y')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
plt.tight_layout()

ax[1].plot(df.y,df.x,'b-o',label='x')
ax[1].legend()
ax[1].set_title('Y vs X')
ax[1].set_xlabel('Y')
ax[1].set_ylabel('X')
plt.tight_layout()
````
