````python
df2=pd.read_csv(r"E:\Dropbox\Study\Python Notes\Main Notes\Datasets\flights.csv")
df2.head(3)
````

````python
row=['month']
col=['year']
value=['passengers']
aggfunc=['sum']
sns.heatmap(df2.pivot_table(value,row,col,aggfunc),annot=False,cmap='viridis',lw=1,linecolor='white')
````

<div style="text-align: center;">
    <h2> Clustermap  </h2>
</div>

````python
row=['month']
col=['year']
value=['passengers']
aggfunc=['sum']
sns.clustermap(df2.pivot_table(value,row,col,aggfunc),annot=False,cmap='Greys',lw=1,linecolor='white',standard_scale=None)
````
