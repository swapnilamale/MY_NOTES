## size

```
du -sh * | sort -hr    # Space taken by each file in a pwd
free -g/m
```

```
# Checking memory allocation in sorted order

ps -eo pid,user,%mem,rss,command --sort -rss | awk '{rss_mb = $4/1024; printf "%-8s %-10s %-8s %-8s %.2fMB\t%s\n", $1, $2, $3, $4, rss_mb, $5}' | head -150

ps -eo user,rss | awk '{sum[$1]+=$2} END {for (u in sum) print u, sum[u]/1024/1024 " GB"}'   # Checking Ram Size
```

## screen

```
screen -ls                        # list of the screen aviailable
screen -r <screen_name>           # Read Screen
screen -S <screen_name>           # Create a screen by of name <screen_name>
screen -X -S <screen_name> quit   # Deleting screen

(Ctrl + a) + d    # Detaching Screen
```

## pip commands

```
pip list                                    : getting the list of all python libraries installed
pip freeze > requirements.txt               : creating requirement.txt for creating python environment (including dependencies)
pip list --format=freeze > requirements.txt : (excluding dependencies)

pip install <library>            : installing python libraries
pip install <library>==<version> : installing perticular version of python library
pip install --upgrade <library>  : upgrading library
pip install -r requirements.txt  : install python libraries from requirements.txt

<library>.__version__   :  Checking the version of <library>    
```
